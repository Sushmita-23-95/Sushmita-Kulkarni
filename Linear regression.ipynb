```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
imp=load_boston()
```

```python
imp
```

```python
Data=pd.DataFrame(imp.data)
Data
```

```python
Data.head()
```

```python
Data.columns=imp.feature_names
```

```python
Data.head()
```

```python
Data['Price']=imp.target
```

```python
Data.shape
```

```python
X=Data.drop('Price',axis=1)
y=Data['Price']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn import linear_model
```

```python
regr=linear_model.LinearRegression().fit(X,y)
```

```python
regr
```

```python
y_preds=regr.predict(X)
y_preds
```

```python
regr.coef_
```

```python
regr.intercept_
```

```python
rss=np.sum((y-y_preds)**2)
rss
```

```python
import sklearn.metrics as metrics
```

```python
r2=metrics.r2_score(y,y_preds)
```

```python
r2
```
